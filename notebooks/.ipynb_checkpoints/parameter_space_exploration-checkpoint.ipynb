```python
import sys
sys.path.append('..')

import src as odyn
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 

import geopandas as gpd
from shapely.geometry import Point, Polygon, MultiPolygon, shape
from pyproj import Proj
```

```python
model = odyn.OpinionNetworkModel(probabilities = [.45,.1,.45])
model.populate_model(num_agents = 1000, show_plot = True)
sim = odyn.NetworkSimulation()
sim.run_simulation(model = model,
           store_results = True
          )
sim.plot_simulation_results()
```

```python
means = []
for result in sim.results:
    means.append(result['belief_df']['belief'].mean(axis = 0))

print(means)
time_steps = range(len(means))
y_interval = np.arange(0,2)
fig, ax = plt.subplots()
ax.set_title('Average belief across timesteps', fontsize=14)
ax.set_xlabel('Time step', fontsize=14)
ax.set_ylabel('Mean belief', fontsize=14)
ax.plot(time_steps,means, color = 'red', marker = 'o')
ax.set_xlim = ((0, max(time_steps)))
ax.set_ylim = (xmin=0, xmax=2)
ax.grid(axis = 'y', visible = True)
plt.show
```
